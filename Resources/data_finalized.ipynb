```python
# Import
import pandas as pd
```

```python
# Original CSV to DataFrame
full_df = pd.read_csv("Listing_of_Active_Businesses.csv")
full_df.head()
```

```python
# Column Renames
full_df = full_df.rename(columns={"BUSINESS NAME": "BUSINESS_NAME", "STREET ADDRESS":"STREET_ADDRESS", "PRIMARY NAICS DESCRIPTION": "DESCRIPTION", "LOCATION": "COORDINATES", "ZIP CODE":"ZIPCODE", "LOCATION START DATE": "STARTDATE"})
# Column reorder
Cleaned_df = full_df[["BUSINESS_NAME", "STREET_ADDRESS", "CITY", "ZIPCODE","NAICS","DESCRIPTION", "STARTDATE","COORDINATES"]]
# Data Error Repaired
Cleaned_df = Cleaned_df.replace("A A OFICINA CENTRAL HISPANA DE LOS ANGELES /C",'A A OFICINA CENTRAL HISPANA DE LOS ANGELES')
```

```python
#Drop Nan
Cleaned_df.dropna(how="any", inplace=True)
Cleaned_df
```

```python
# Seperating the Naics into just the Sector level
import math

def first_two(d):                
     return (d // 10 ** (int(math.log(d, 10)) - 1))

Cleaned_df['NAICS_SECTOR'] = Cleaned_df.NAICS.apply(first_two)
```

```python
# Changing the Start Date just to year
Cleaned_df['STARTDATE'] = Cleaned_df['STARTDATE'].str.strip().str[-4:]
```

```python
# Removing () and spliting coordinates into lat and lng
Cleaned_df['COORDINATES'] = Cleaned_df['COORDINATES'].str.replace('(', '',regex=True)
Cleaned_df['COORDINATES'] = Cleaned_df['COORDINATES'].str.replace(')', '',regex=True)
Cleaned_df[['Lat', 'Lng']] = Cleaned_df.COORDINATES.str.split(",", expand = True)
# Seperating Zipcode to just district area
Cleaned_df[['ZIPCODE1', 'Place Zip']] = Cleaned_df.ZIPCODE.str.rsplit("-",n=1, expand = True)
```

```python
# Dropping Duplicates
Cleaned_df.drop_duplicates()
```

```python
# Fixing lat lng datatypes
Cleaned_df["Lat"] = pd.to_numeric(Cleaned_df["Lat"])
Cleaned_df["Lng"] = pd.to_numeric(Cleaned_df["Lng"])
Cleaned_df.dtypes
```

```python
# Making City names matching format
Cleaned_df['CITY'] = Cleaned_df.loc[:,'CITY'].str.upper()
Cleaned_df
```

```python
# Make all Sectors that are the same but with different numbers the same number for visual ease
Cleaned_df = Cleaned_df.replace(32.0,31.0)
Cleaned_df = Cleaned_df.replace(33.0,31.0)
Cleaned_df = Cleaned_df.replace(45.0,44.0)
Cleaned_df = Cleaned_df.replace(49.0,48.0)
```

```python
Cleaned_df = Cleaned_df[["BUSINESS_NAME", "STREET_ADDRESS", "CITY", "ZIPCODE1","NAICS","DESCRIPTION", "NAICS_SECTOR", "STARTDATE","Lat","Lng"]]
Cleaned_df
```

```python
# Dropping non LA data
Cleaned_df.drop(Cleaned_df[Cleaned_df['CITY'] != "LOS ANGELES"].index, inplace = True)
Cleaned_df.drop(Cleaned_df[Cleaned_df['ZIPCODE1'] <= "90000"].index, inplace = True)
```

```python
# Building Sector Dataframe
d={"NAICS_SECTOR":[11.0,21.0,22.0,23.0,31.0,42.0,44.0,48.0,51.0,52.0,53.0,54.0,55.0,56.0,61.0,62.0,71.0,72.0,81.0,92.0],
                              "INDUSTRY_DESCRIPTION":["Agriculture, Forestry, Fishing and Hunting","Mining, Quarrying, and Oil and Gas Extraction",
                                                     "Utilities","Construction","Manufacturing", "Wholesale Trade",
                                                     "Retail Trade","Transportation and Warehousing",
                                                     "Information", "Finance and Insurance", "Real Estate and Rental and Leasing", "Professional, Scientific, and Technical Services",
                                                     "Management of Companies and Enterprises", "Administrative and Support and Waste Management and Remediation Services",
                                                     "Educational Services","Health Care and Social Assistance","Arts, Entertainment, and Recreation",
                                                     "Accommodation and Food Services","Other Services (except Public Administration)","Public Administration"]}
Sector_df = pd.DataFrame(data=d)
Sector_df
```

```python
# Joining information
merged_df = pd.merge(Cleaned_df, Sector_df, on='NAICS_SECTOR', how='inner')
# Reordering of DataFrame
Final_df = merged_df[["BUSINESS_NAME", "STREET_ADDRESS", "CITY", "ZIPCODE1", "NAICS_SECTOR", "INDUSTRY_DESCRIPTION", "STARTDATE","Lat","Lng"]]
Final_df.head()
```

```python
# Export Table to CSV
Final_df.to_csv("LABusinessData_Final.csv", index=False)
```
