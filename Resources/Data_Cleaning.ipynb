```python
# Import
import pandas as pd
```

```python
# Original CSV to DataFrame
full_df = pd.read_csv("Listing_of_Active_Businesses.csv")
full_df.head()
```

```python
# Column Renames
full_df = full_df.rename(columns={"PRIMARY NAICS DESCRIPTION": "DESCRIPTION", "LOCATION": "COORDINATES", "ZIP CODE":"ZIPCODE"})
```

```python
# Column reorder
Cleaned_df = full_df[["BUSINESS NAME", "STREET ADDRESS", "CITY", "ZIPCODE","NAICS","DESCRIPTION", "LOCATION START DATE","COORDINATES"]]
# Data Error Repaired
Cleaned_df = Cleaned_df.replace("A A OFICINA CENTRAL HISPANA DE LOS ANGELES /C",'A A OFICINA CENTRAL HISPANA DE LOS ANGELES')
```

```python
#Drop Nan
Cleaned_df.dropna(how="any", inplace=True)
Cleaned_df
```

```python
# Seperating the Naics into just the Sector level
import math

def first_two(d):                
     return (d // 10 ** (int(math.log(d, 10)) - 1))

Cleaned_df['NAICS SECTOR'] = Cleaned_df.NAICS.apply(first_two)
Cleaned_df.head()
```

```python
# Removing () and spliting coordinates into lat and lng
Cleaned_df['COORDINATES'] = Cleaned_df['COORDINATES'].str.replace('(', '',regex=True)
Cleaned_df['COORDINATES'] = Cleaned_df['COORDINATES'].str.replace(')', '',regex=True)
Cleaned_df[['Lat', 'Lng']] = Cleaned_df.COORDINATES.str.split(",", expand = True)
```

```python
# Dropping Duplicates
Cleaned_df.drop_duplicates()
```

```python
# Seperating Zipcode to just district area
Cleaned_df[['ZIPCODE1', 'Place Zip']] = Cleaned_df.ZIPCODE.str.rsplit("-",n=1, expand = True)
Cleaned_df
```

```python
# Fixing lat lng datatypes
Cleaned_df["Lat"] = pd.to_numeric(Cleaned_df["Lat"])
Cleaned_df["Lng"] = pd.to_numeric(Cleaned_df["Lng"])
Cleaned_df.dtypes
```

```python
# Making City names matching format
Cleaned_df['CITY'] = Cleaned_df.loc[:,'CITY'].str.upper()
Cleaned_df
```

```python
# Adding an ID column for SQL use
Cleaned_df['ID'] = Cleaned_df.index
Cleaned_df = Cleaned_df[["ID","BUSINESS NAME", "STREET ADDRESS", "CITY", "ZIPCODE1","NAICS","DESCRIPTION", "NAICS SECTOR", "LOCATION START DATE","COORDINATES","Lat","Lng"]]
Cleaned_df
```

```python
# Tables to CSV
Sector_df.to_csv("sectors.csv", index=False)
Cleaned_df.to_csv("BusinessData_cleaned.csv", index=False)
```

```python
# Joining information
merged_df = pd.merge(Cleaned_df, Sector_df, on='NAICS SECTOR', how='inner')
merged_df
```
